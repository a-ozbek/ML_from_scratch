# Linear Regression

```python
import numpy as np
RANDOM_SEED = 43
np.random.seed(RANDOM_SEED)
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X, y = datasets.make_regression(n_features=1, n_samples=200, bias=100, noise=20)
X, y = np.matrix(X), np.matrix(y.reshape((len(X), 1)))
```

```python
# Parameters
N_iterations = 100
learning_rate = 0.001

# Randomly initialize w vector
w = np.random.rand(2, 1)
w = np.matrix(w)

# Augment X
X = np.concatenate([X, np.ones((X.shape[0], 1))], axis=1)

# Iterations
for i in range(N_iterations):
    w_gradient = 2 * X.T * (X * w - y)
    w = w - w_gradient * learning_rate
    print 'i:', i
    print 'w:', w
```

```python
# Plot the results
fig = plt.figure()
# Points
plt.scatter(X[:,0], y, c='b');
plt.xlabel('x'); plt.ylabel('y')
# Plot prediction
end_points = np.array([np.min(X[:,0]), np.max(X[:,0])])
f = lambda x: x * w[0,0] + w[1,0]
plt.plot(end_points, f(end_points), c='r')
```
